```python
import numpy as np
```

```python
potentials = np.random.random((507, 3))
energy_diff = np.empty([507, 3, 3])
states = 3
for i in range(states):
    energy_diff[:, i] = (potentials.T - potentials[:, i]).T

print(energy_diff.shape)
```

```python
coupling_structures = []
for i in range(3):
    state = i
    count = 0
    coupling_structure_per_state = []
    for j, energies in enumerate(energy_diff[:, state]):
        energies = np.delete(energies, state, 0)
        indices = np.argwhere(abs(energies) < 0.2)
        if len(indices) > 0:
        # print("contains coupling region")
            coupling_structure_per_state.append(j)
            count += 1
    coupling_structures.append(coupling_structure_per_state)

    # print(count)
print(coupling_structures)
# for i in range(states):
#     for energies in energy_diff:
#         print(energies.shape)
        # check_energies = np.delete(energies, i, 0)
        # print(check_energies.shape)
        # break
```
